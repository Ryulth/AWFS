```python
import pandas as pd
```

```python
df=pd.read_csv('./datas/khu_courses.csv')
```

```python
df.columns
```

```python
df=df[["학수번호-분반",'강좌명','강의시간/강의실']]
df
```

```python
df=df[df['학수번호-분반'].str.contains("CSE|SWC")]
df
```

```python
weekday=['월','화','수','목','금','토','일']
def get_details(s):
    s=s.split(')')
    day=''
    time=''
    room=''
    for i in s:
        i=i.replace('(','')
        k=i.split(' ')
        if k[0] in weekday:
            day+=k[0]+','
            time+=k[1]+','
            room+=k[2]+','
    return day,time,room
```

```python
df['day']= df.apply(lambda x:get_details(x['학수번호-분반'])[0], axis=1)
df['time']= df.apply(lambda x:get_details(x['학수번호-분반'])[1], axis=1)
df['room'] = df.apply(lambda x:get_details(x['학수번호-분반'])[2], axis=1)
```

```python
df['day']= df['학수번호-분반'].map(lambda x:get_details(x)[0])
```
